> - converting zeros to NaN

```python
df_clean.revenue=df_clean.revenue.replace(0,np.nan)
```

```python
df_clean.budget= df_clean.budget.replace(0,np.nan)
```

```python
df_clean.budget
```

```python
df_clean.revenue
```

##### release_date is an object type
> - converting release_date to datetime instead of string

```python
df_clean['release_date'] = pd.to_datetime(df_clean['release_date'])
```

```python
df_clean['release_date'].head()
```

##### one duplicated row
> - reomving it

```python
df_clean.drop_duplicates(inplace=True)
```

## Exploratory Data Analysis

### Research Question 1 (What is the most popular genre from year to year?)

```python
def most_popular(df,popular):
    popular_per_year=df.groupby(['release_year',popular]).popularity.sum()
    popular_dict = {}
    for i in range(1960,2016):
        key=str(i)+': '+popular_per_year[i].idxmax()
        value= popular_per_year[i].max()
        popular_dict[key] = popular_dict.get(key,0)+value
        
    return popular_dict

       
def plot_graph(x,y,title,x_label,y_label,fonts,graph_type):
    if graph_type=='barh':
        plt.barh(x,y,color='midnightblue')
    elif graph_type =='bar':
        plt.bar(x,y,color='midnightblue')
    else:
        plt.scatter(x,y)
    plt.title(title,fontsize=fonts[0],fontweight='bold')
    plt.xlabel(x_label,fontsize=fonts[1])
    plt.ylabel(y_label,fontsize=fonts[1]);
    
```

```python
genres_dict = most_popular(df_clean,'genres')
```

```python
plt.figure(figsize=(15,20))
plot_graph(list(genres_dict.keys()),list(genres_dict.values()),'Most popular genres through years',
          'Popularity','genres per year',[16,14],'barh')
```

As shown in the graph the most popular genres are Action|Adventure|Science Fiction|Thriller in 2015 followed by Drama in 2014 followed by Comedy in 2013

### Research Question 2  (Who is the most popular actor from year to year?)

```python
df_sub= df_clean[['id','cast']]
df_sub.index=df_sub.id
```

```python
df_sub= df_sub.cast.str.split('|',expand=True)
df_sub= df_sub.stack()
```

```python
df_sub= pd.DataFrame(df_sub)
```

```python
df_sub.rename(columns={0:'cast_adj'},inplace=True)
```

```python
df= pd.merge(df_sub,df_clean,on='id')
```

```python
actors_dict=most_popular(df,'cast_adj')
```

```python
plt.figure(figsize=(15,20))
plot_graph(list(actors_dict.keys()),list(actors_dict.values()),'Most popular actors through years',
          'Popularity','Actor per year',[16,14],'barh')
```

from the bar graph we can see that most popular actor is Tom Hardy in 2015 followed by jessica Chastian in 2014 followed by Orlando Bloom in 2003

### Research Question 3  (what is the movie with highest profit?)

```python
df_clean['profit']= df_clean['revenue']-df_clean['budget']
```

```python
df_prof_above_500m= df_clean[df_clean.profit>500000000] 
movies_profit= df_prof_above_500m.groupby('original_title').profit.sum()
```

```python
movies_profit.sort_values(ascending=False,inplace=True)
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
plot_graph(movies_profit.index[:10],movies_profit.values[:10],'Movies profit',
          'Movies','profit',[16,12],'bar')
```

from graph we can see that Avatar has the highest profit with 2.847  billion followed  by  Star Wars: The Force Awakens with 2.064 billion followed by Titanic with 2.207 billion

### Research Question 4   (What is the year with highest and lowest profit?)

```python
profit_per_year= df_clean.groupby('release_year').profit.sum()
```

```python
high_year=str( profit_per_year.idxmax())
high_profit=str(profit_per_year.max())
low_year= str(profit_per_year.idxmin())
low_profit=str(profit_per_year.min())
print('Year with highest profit is '+high_year+' with '+high_profit)
print('Year with lowest profit is '+low_year+' with '+low_profit)
```

### Research Question 5   (who is the director with highest movie profit and what are his top 5 movies?)

From the treemap above we can see James Cameron top 5 movies having Avatar in the first place followed by Titanic

### Research Question 7  (which production company have the highest movie budget and highest profit?)

```python
df_company= df_clean[['id','production_companies']]
```

```python
df_company.index= df_company.id
```

```python
df_company= df_company.production_companies.str.split('|',expand=True)
```

```python
df_company= df_company.stack()
```

```python
df_company= pd.DataFrame(df_company)
```

```python
df_company.rename(columns={0:'production_company'},inplace=True)
```

```python
df_company_adj= pd.merge(df_company,df_clean,on='id')
```

```python
company_with_highest_profit= df_company_adj.groupby('production_company').profit.max()
company_with_highest_budget= df_company_adj.groupby('production_company').budget.max() 
```

```python
company_with_highest_budget.sort_values(ascending=False,inplace=True)
company_with_highest_profit.sort_values(ascending=False,inplace=True)
```

```python
company_with_highest_budget= company_with_highest_budget.iloc[:10]
company_with_highest_profit= company_with_highest_profit.iloc[:10]
```

```python
#top 10 production companies with the highest budget
plt.figure(figsize=(15,9))
plt.xticks(rotation=45);
plot_graph(company_with_highest_budget.index,company_with_highest_budget.values,'Top 10 Production Companies with the highest budget',
          'Production Companies','Budget',[15,12],'bar')
```

From the Bar graph we can see that Boram Entertainment inc is the copmany with the highest budget

```python
#top 10 production companies with the highest profit
plt.figure(figsize=(15,9))
plt.xticks(rotation=45);
plot_graph(company_with_highest_profit.index,company_with_highest_profit.values,'Top 10 Production Companies with the highest profit',
          'Production Companies','profit',[15,12],'bar')
```

```python
company_with_highest_profit
```

From the Bar graph we can see that we have four companies with the highest profit and they are Ingenious Film Partners, Twentieth Century Fox Film Corporation, Lightstorm Entertainment, and Dune Entertainment with 2.55 billion
