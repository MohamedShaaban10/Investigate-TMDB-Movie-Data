```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
from matplotlib.ticker import PercentFormatter as pct
sns.set()
```

```python
df_cars= pd.read_csv('Bar Chart/bar_chart_data.csv')
df_cars.head()
```

```python
plt.figure(figsize=(10,6))

plt.bar(df_cars.Brand,df_cars.loc[:,'Cars Listings'],color='midnightblue')
plt.xticks(rotation=45)
plt.yticks(fontsize=12)
plt.title('Cars Listings by Brand',fontsize=16,fontweight='bold')
plt.ylabel('Number of Listings');
```

```python
plt.figure(figsize=(10,6))

plt.barh(df_cars.Brand,df_cars.loc[:,'Cars Listings'],color='midnightblue')
plt.yticks(rotation=45)
plt.xticks(fontsize=12)
plt.title('Cars Listings by Brand',fontsize=16,fontweight='bold')
plt.xlabel('Number of Listings');
```

```python
df_fuel= pd.read_csv('../Pie Chart/pie_chart_data.csv')
df_fuel
```

```python
plt.figure(figsize=(10,6))
plt.pie(df_fuel.loc[:,'Number of Cars'],labels=df_fuel.loc[:,'Engine Fuel Type'],
        autopct='%.2f%%',textprops={'size':'x-large',
                                     'fontweight':'bold',
                                     'rotation':'30',
                                     'color':'w'});
plt.legend()
plt.title('Cars by Engine Type',fontsize=18,fontweight='bold');
```

```python
df_fuel= pd.read_csv('../Stacked Area Chart/stacked_area_chart_data.csv')
df_fuel.head()
```

```python
colors=['#011638','#7e2987','#ef2026']
labels=['Diesel','Petrol','Gas']
sns.set_style('white')

plt.figure(figsize=(12,6))
plt.stackplot(df_fuel.Year,
              df_fuel.Diesel,
              df_fuel.Petrol,
              df_fuel.Gas,colors=colors,
             edgecolor='none')
plt.legend(labels=labels)
plt.ylabel('Number Of Cars',fontsize=13)
plt.title('Popularity of Engine Fuel Types from (1982,2016)',fontsize=14,fontweight='bold')
plt.xticks(df_fuel.Year,rotation=45)
sns.despine();
```

```python
df_rev= pd.read_csv('../Line Chart/line_chart_data.csv')
df_rev.head()
```

```python
df_rev['Date']=pd.to_datetime(df_rev['Date'])
```

```python
df_rev.head()
```

```python
labels=['S&P 500','FTSE 100']
sns.set()
plt.figure(figsize=(20,8))
plt.plot(df_rev.Date,df_rev.GSPC500)
plt.plot(df_rev.Date,df_rev.FTSE100)
plt.title('S&P vs FTSE Returns (2000-2010)',fontsize=14,fontweight='bold')
plt.legend(labels=labels, fontsize='large')
sns.despine()
plt.ylabel('Returns%',fontsize=12)
plt.xlabel('Date in years',fontsize=12); 
```

```python
df_h2_2008= df_rev[((df_rev.Date>='2008-07-01') & (df_rev.Date<='2008-12-31'))]
```

```python
df_h2_2008
```

```python
labels=['S&P 500','FTSE 100']
sns.set()
plt.figure(figsize=(20,8))
plt.plot(df_h2_2008.Date,df_h2_2008.GSPC500,color='darkblue')
plt.plot(df_h2_2008.Date,df_h2_2008.FTSE100,color='darkred')
plt.title('S&P vs FTSE Returns (H2 2008)',fontsize=14,fontweight='bold')
plt.legend(labels=labels, fontsize='large')
sns.despine()
plt.ylabel('Returns%',fontsize=12)
plt.xlabel('Date in (H2 2008)',fontsize=12); 
```

```python
df_prices= pd.read_csv('../Scatter Plot/scatter_data.csv')
df_prices.head()
```

```python
df=df_prices.loc[:,['Area (ft.)','Price']]
```

```python
df.head()
```

```python
sns.set()
plt.figure(figsize=(12,8))
scatter=plt.scatter(df['Area (ft.)'],df.Price,alpha=0.6,c=df_prices['Building Type'],cmap='mako')
plt.title('Relationship between Price and Area',fontsize=14,fontweight='bold')
plt.legend(*scatter.legend_elements(),title='Building Type')
plt.xlabel('Area')
plt.ylabel('Price');
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=df['Area (ft.)'],y=df.Price,hue=df_prices['Building Type'],
                 palette=['black','darkblue','purple','pink','white'],s=100)
plt.title('Relationship between Price and Area',fontsize=14,fontweight='bold')
plt.legend(*scatter.legend_elements(),title='Building Type')
plt.xlabel('Area')
plt.ylabel('Price');
```

```python
df_sales= pd.read_csv('Combo Plot/scatter_plot_ii.csv')
df_sales.head()
df_sales.Budget.max()
```

```python
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x='Budget',y='Sales',data=df_sales,scatter_kws={'color':'k'},line_kws={'color':'red'})
plt.xticks(np.arange(50,500,50))
plt.title('Relationship Between Budget vs Sales',fontsize=14,fontweight='bold');
```

```python
sns.lmplot(x='Budget',y='Sales',data=df_sales,
           scatter_kws={'color':'k'},line_kws={'color':'red'},height=8)
plt.xticks(np.arange(50,500,50))
plt.title('Relationship Between Budget vs Sales',fontsize=14,fontweight='bold');
```

```python
df_insurance=pd.read_csv('Combo Plot/regression_auto_insurance_sweden.csv')
df_insurance.head()
df_insurance.Payment.max()
```

```python
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x = "Claims",  y = "Payment", 
            data=df_insurance,scatter_kws={'color':'k'},line_kws={'color':'red'})
plt.xticks(np.arange(0,180,20))
plt.title('Effect of Payment on Claims',fontsize=14,fontweight='bold');
```

```python
df_kd_nuggets= pd.read_csv('Combo Plot/Bar and line Chart/bar_line_chart_data.csv')
df_kd_nuggets.head()
```

```python
sns.set_style('white')
fig,ax=plt.subplots(figsize=(10,7))
ax.bar(df_kd_nuggets.Year,df_kd_nuggets.Participants,color='k')
ax.set_ylabel('Number Of Participants')
axl= ax.twinx()
axl.set_ylabel('Python Users %',color='#b60000')
axl.tick_params(axis='y',colors='#b60000')
axl.set_ylim(0,1)
axl.yaxis.set_major_formatter(pct(xmax=1.0))
axl.plot(df_kd_nuggets.Year,df_kd_nuggets['Python Users'],color='#b60000',marker='D')
ax.set_title('KD Nuggets Survey Python Users (2012-2019)',fontsize=14,fontweight='bold');
ax.set_xlabel('Year');
```
